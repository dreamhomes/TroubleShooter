## EWMA

Simple moving average, Exponential weighted moving average.

```python
import sys

sys.path.append("..")

import blot

import numpy as np
import pandas as pd

from datetime import datetime

from troubleshooter.anomaly.tseries.move_average import move_average, ewma
```

```python
data = pd.read_csv("../data/storage.csv")
data["time"] = data.time.apply(lambda ts: datetime.fromtimestamp(ts))
# data = data.set_index(["time"])
data = data[(data["time"] > datetime(2020, 8, 30)) & (data["time"] < datetime(2020, 9, 14))]
```

```python
graph = data.blot(x='time', y=['value'], colors={'value': 'blue'}, dimensions=(None, 900))


with open("../output/value.json", "w") as fw:
    fw.write(graph.to_json())
```

```python
data["EWMA"] = ewma(data["value"].values)
data["MA"] = move_average(data["value"].values)
data["E_MA"] = data["value"].values - move_average(data["value"].values)
data["E_EWMA"] = data["value"].values - ewma(data["value"].values)

data["upper"] = data["value"].rolling(window=10).max() # 滑动平均取上界
data["lower"] = data["value"].rolling(window=10).min() # 滑动平均取下界

data.head()
```

```python
data.blot(x='time', y=['value', 'EWMA', 'MA','E_MA', 'E_EWMA'], dimensions=(1400, 900), colors={'value': 'blue', 'EWMA': 'green', 'E_MA': 'purple', 'E_EWMA': 'pink', 'MA': 'maroon'}, upper='upper', lower='lower', anomaly_color='rgba(0,0,0,0)')
```
